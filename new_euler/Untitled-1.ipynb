```python

def isPrime(n):
    if(n%2 == 0):
         return False
    for i in range(3, int(n**(1/2)) + 1, 2):
        if(n%i == 0):
            # print(i)
            return False
    return True
```

```python
def SieveOfEratosthenes(n):
    
    prime = [True for i in range(n+1)]
    p=2 
    while p**2 <= n:
        if(prime[p]):
            for i in range(p*p, n+1, p):
                prime[i] = False
        p+=1
    returnArr=[]
    for p in range(3, n+1):
        if(prime[p]):
            returnArr.append(p)
    return returnArr
```

```python
def isConcatPrime(a,b):
    return isPrime(int(str(a) + str(b))) and isPrime(int(str(b) + str(a)))
# [557, 607, 613, 617, 673]
# print(isConcatPrime(607, 673))

primeList = SieveOfEratosthenes(1000)
primeLeng = len(primeList)
result = 0;
```

```python
def subArray(arr, lowerLimit, upperLimit):
    newArr=[]
    for i in range(lowerLimit, upperLimit+1):
        try:
            newArr.append(arr[i])
        except IndexError:
            return newArr

    return newArr
print(subArray([1,2,3,4,5], 2, 2))
```
